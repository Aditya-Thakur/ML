```python
def change_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
        if Pclass==1:
            cols[0]=38
            return cols[0]
        if Pclass==2:
            cols[0]=30
            return cols[0]
        if Pclass==3:
            cols[0]=25
            return cols[0]
    else:
        return Age
```

```python
titanic['Age']=titanic[['Age','Pclass']].apply(change_age, axis=1)
```

```python
titanic['Age']
```

```python
X=titanic.iloc[:,2:].values
y=titanic.iloc[:,1].values
```

```python
#5.Handling categorical data
# country and purchased are categorical data, we are going to replace it with numeric data becoz only numeic data is processed
from sklearn.preprocessing import LabelEncoder
le_X=LabelEncoder()  #no para needed,le_X =object of x


#ohe=OneHotEncoder(categorical_features=[4])#column on which (0) we are applying categorical features
X[:,1]=le_X.fit_transform(X[:,1])
X   
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=601)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
